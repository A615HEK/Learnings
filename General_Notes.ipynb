# Terminology
## Eval Metrics
Accuracy = $\frac{Correct\  predictions}{All\ predictions}$ -- Correctness<br><br>
Precision = $\frac{TP}{TP+FP}$ --> P(Positive / Predicted+) -- Consistency<br><br>
Recall = $\frac{TP}{TP+FN}$ --> P(Positive / Actual+)<br><br>
F1 Score = 2 $\frac{Precision\ *\ Recall}{Precision\ +\ Recall}$
